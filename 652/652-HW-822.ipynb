```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/40515/Desktop/spine.csv')
df.head()
```

```python
features =df[df.columns[0:12]]
labels=df["Class_att"].values
from sklearn.cross_validation import train_test_split
x=features
y=labels
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=0)
train_x
```

```python
train_y
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_x, train_y)

from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test_y, model.predict_proba(test_x)[:,1], pos_label=1)
plt.plot(fpr, tpr, color="green")
print("LR auc = ", metrics.auc(fpr, tpr))
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train_x, train_y)

from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test_y, gnb.predict_proba(test_x)[:,1], pos_label=1)
plt.plot(fpr, tpr, color="green")
print("Naive Bayes auc = ", metrics.auc(fpr, tpr))
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
kNN = KNeighborsClassifier(n_neighbors=6)
kNN.fit(train_x, train_y) 

from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(test_y, kNN.predict_proba(test_x)[:,1], pos_label=1)
plt.plot(fpr, tpr, color="green")
print("kNN auc = ", metrics.auc(fpr, tpr))
plt.show()
```
