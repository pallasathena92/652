```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('E:/stevens/652multipul_analytics/data/binary.csv')
df.head()
```

```python
import statsmodels.api as sm
labels = df["admit"].values
features = df[["gre","gpa","rank"]]
ones=np.transpose([np.repeat(1,features.shape[0])])
#need to add constant term since Logit does not automatically add it!!
features2 = np.concatenate((ones, features), axis = 1) 
model = sm.Logit(labels, features2)
result = model.fit() #can specify method, e.g., method="bfgs"
result.summary()
```

```python
from sklearn.cross_validation import train_test_split
x=features
y=labels
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=0)
train_x
```

```python
train_y
```

```python
labels = train_y
features = train_x
ones=np.transpose([np.repeat(1,features.shape[0])])
#need to add constant term since Logit does not automatically add it!!
features2 = np.concatenate((ones, features), axis = 1) 
model = sm.Logit(labels, features2)
result = model.fit() #can specify method, e.g., method="bfgs"
result.summary()
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(labels, result.predict(features2), pos_label=1)
plt.plot(fpr,tpr, color="blue")
print("auc for train = ", metrics.auc(fpr, tpr))
plt.show()
```

```python
labels = test_y
features = test_x
ones=np.transpose([np.repeat(1,labels.shape[0])])
#need to add constant term since Logit does not automatically add it!!
features2 = np.concatenate((ones, features), axis = 1) 
model = sm.Logit(labels, features2)
result = model.fit() #can specify method, e.g., method="bfgs"
result.summary()
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(labels, result.predict(features2), pos_label=1)
plt.plot(fpr,tpr, color="blue")
plt.ylabel("TPR"); plt.xlabel("FPR");
print("auc for test = ", metrics.auc(fpr, tpr))
plt.show()
```

```python
prec, recall, thresholds = metrics.precision_recall_curve(labels,result.predict(features2), pos_label=1)
plt.plot(recall, prec, color="blue")
plt.show()
```
